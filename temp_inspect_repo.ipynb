```python
import shutil
from pathlib import Path

repo_root = Path('/workspaces/Amethyst')
hynis_root = Path('/workspaces/Hynis')

# Copy patchsvc source and manifest
src_patchsvc = hynis_root / 'JavaApp/src/patchsvc'
dst_patchsvc = repo_root / 'JavaApp/src/patchsvc'
if dst_patchsvc.exists():
    shutil.rmtree(dst_patchsvc)
shutil.copytree(src_patchsvc, dst_patchsvc)
shutil.copy2(hynis_root / 'JavaApp/src/patchsvc.txt', repo_root / 'JavaApp/src/patchsvc.txt')

# Copy svc native libs
src_svc = hynis_root / 'Natives/resources/Frameworks/svc'
dst_svc = repo_root / 'Natives/resources/Frameworks/svc'
if dst_svc.exists():
    shutil.rmtree(dst_svc)
shutil.copytree(src_svc, dst_svc)

print('Copied patchsvc and svc files')

# Update JavaApp/Makefile
makefile_path = repo_root / 'JavaApp/Makefile'
makefile_text = makefile_path.read_text()
if '$(OUTPUTDIR)/patchsvc.jar: $(CLASSES)' not in makefile_text:
    insertion = """
$(OUTPUTDIR)/patchsvc.jar: $(CLASSES)
	@set -e
	@mv $(basename $@)/de/maxhenkel/lame4j/Mp3Decoder.class $(basename $@)/de/maxhenkel/lame4j/Mp3Decoder.class.patch
	@mv $(basename $@)/de/maxhenkel/lame4j/Mp3Encoder.class $(basename $@)/de/maxhenkel/lame4j/Mp3Encoder.class.patch
	@mv $(basename $@)/de/maxhenkel/opus4j/OpusDecoder.class $(basename $@)/de/maxhenkel/opus4j/OpusDecoder.class.patch
	@mv $(basename $@)/de/maxhenkel/opus4j/OpusEncoder.class $(basename $@)/de/maxhenkel/opus4j/OpusEncoder.class.patch
	@mv $(basename $@)/de/maxhenkel/rnnoise4j/Denoiser.class $(basename $@)/de/maxhenkel/rnnoise4j/Denoiser.class.patch
	@mv $(basename $@)/de/maxhenkel/speex4j/AutomaticGainControl.class $(basename $@)/de/maxhenkel/speex4j/AutomaticGainControl.class.patch
	@rm -rf $(basename $@)/de/maxhenkel/*/*.class
	@echo "Creating $@"
	@$(BOOTJDK)/jar -cfm $@ $(SOURCEDIR)/patchsvc.txt -C $(basename $@) .
"""
    replace_target_line = """
$(OUTPUTDIR)/patchjna_agent.jar: $(CLASSES)
	@set -e
	@mv $(basename $@)/com/sun/jna/Platform.class $(basename $@)/com/sun/jna/Platform.class.patch
	@echo "Creating $@"
	@$(BOOTJDK)/jar -cfm $@ $(SOURCEDIR)/patchjna_agent.txt -C $(basename $@) .
"""
    makefile_text = makefile_text.replace(replace_target_line, replace_target_line + insertion)
    makefile_path.write_text(makefile_text)
    print('Updated JavaApp/Makefile with patchsvc target')
else:
    print('JavaApp/Makefile already contains patchsvc target')

# Update JavaLauncher.m with patchsvc agent line
javalauncher_path = repo_root / 'Natives/JavaLauncher.m'
jtext = javalauncher_path.read_text()
if 'video.fix_simple_voice_chat_mod' not in jtext:
    marker = '    if(getPrefBool(@"general.cosmetica")) {\n        margv[++margc] = [NSString stringWithFormat:@"-javaagent:%@/arc_dns_injector.jar=23.95.137.176", librariesPath].UTF8String;\n    }\n'
    replacement = marker + '    if(getPrefBool(@"video.fix_simple_voice_chat_mod")) {\n        margv[++margc] = [NSString stringWithFormat:@"-javaagent:%@/patchsvc.jar=", librariesPath].UTF8String;\n    }\n'
    if marker in jtext:
        jtext = jtext.replace(marker, replacement)
        javalauncher_path.write_text(jtext)
        print('Updated Natives/JavaLauncher.m with patchsvc injection')
    else:
        raise RuntimeError('Could not find insertion marker in JavaLauncher.m')
else:
    print('Natives/JavaLauncher.m already contains patchsvc injection')

# Add preference default
prefs_path = repo_root / 'Natives/PLPreferences.m'
prefs_text = prefs_path.read_text()
if '"fix_simple_voice_chat_mod": @NO' not in prefs_text:
    prefs_text = prefs_text.replace('"allow_microphone": @NO\n        }.mutableCopy,', '"fix_simple_voice_chat_mod": @NO,\n            "allow_microphone": @NO\n        }.mutableCopy,')
    prefs_path.write_text(prefs_text)
    print('Added default preference video.fix_simple_voice_chat_mod')
else:
    print('Preference default already exists')

# Add preference UI switch
ui_path = repo_root / 'Natives/LauncherPreferencesViewController.m'
ui_text = ui_path.read_text()
if '"video.fix_simple_voice_chat_mod"' not in ui_text:
    ui_text = ui_text.replace('            {@"key": @"allow_microphone",\n              @"hasDetail": @YES,\n              @"icon": @"mic",\n              @"type": self.typeSwitch\n            },\n        ], @[\n',
        '            {@"key": @"allow_microphone",\n              @"hasDetail": @YES,\n              @"icon": @"mic",\n              @"type": self.typeSwitch\n            },\n            {@"key": @"fix_simple_voice_chat_mod",\n              @"hasDetail": @YES,\n              @"icon": @"waveform",\n              @"type": self.typeSwitch\n            },\n        ], @[\n')
    ui_path.write_text(ui_text)
    print('Added UI toggle for fix_simple_voice_chat_mod')
else:
    print('UI toggle already exists')

# Add English localization entries
loc_path = repo_root / 'Natives/resources/en.lproj/Localizable.strings'
loc_text = loc_path.read_text()
if 'preference.title.fix_simple_voice_chat_mod' not in loc_text:
    loc_text += '\n"preference.title.fix_simple_voice_chat_mod" = "Fix Simple Voice Chat";\n"preference.detail.fix_simple_voice_chat_mod" = "Apply compatibility patches for Simple Voice Chat mods.";\n'
    loc_path.write_text(loc_text)
    print('Added English localization strings for fix_simple_voice_chat_mod')
else:
    print('Localization strings already present')
```
